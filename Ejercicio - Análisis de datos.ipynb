Éstos corresponden a los contagios diarios de COVID-19 a partir del 22 de marzo del 2020 al 6 de agosto del 2021.

Ordene los datos en un DataFrame de pandas correspondiendo fecha con numero de contagios.

```python
start = "2020-03-22"
end = "2021-08-06"
date = pd.date_range(start, end)

separacion = np.linspace(1, len(D), num = len(D))
tabla = pd.DataFrame(columns = ["Fecha", "Contagios"], index = separacion)

tabla["Fecha"] = date
tabla["Contagios"] = D

tabla 
```

Calcule la variación relativa día a día que mejor se aproxime a una distrubución normal

```python
r = ((tabla["Contagios"] - tabla["Contagios"].shift()) / tabla["Contagios"].shift())
r = r.replace(np.inf, np.nan).dropna()

mean = r.mean()
desvesta = r.std()
mean, desvesta
```

```python
def rendimientos(mean, desvesta, start, dias, separacion) :
    sim = np.random.normal(mean, desvesta, size = (dias, separacion))
    date =pd.date_range(start, periods = dias)
    return pd.DataFrame(data = sim, index = date)
```

```python
simulacion = rendimientos(mean, desvesta, start, 365, 10000)
simulacion
```

Simule 10000 escenarios para el siguiente año y calcule la probabilidad de que los contagios sean menores que 50.

### Para el siguiente año

```python
rend2 = rendimientos(mean, desvesta, "2022-01-01", 365, 10000)
rend_acum = (rend2 + 1).cumprod()
val_sim = tabla["Contagios"].iloc[-1] * rend_acum
```

```python
valor_deseado = tabla["Contagios"].iloc[-1] * 1.5
sobrepasa = (val_sim < valor_deseado)
sobrepasa_prob = sobrepasa.mean(axis = 1)
plt.figure()
plt.plot(sobrepasa_prob)
```
